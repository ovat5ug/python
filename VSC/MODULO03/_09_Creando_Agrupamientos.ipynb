```python
datos = pd.read_csv('alquiler_residencial_filtradoSinNulos.csv', sep = ';')
```

```python
barrios = ['Ate','Barranco','Comas','Lince','El Agustino','San Luis','Callao']
seleccion = datos['Distrito'].isin(barrios)#".isin" significa que esten dentro de 
datos = datos[seleccion]
```

```python
datos['Distrito'].drop_duplicates()#eliminamos los duplicados
```

```python
grupo_barrio = datos.groupby('Distrito')#por cual columna se quiere agrupar
```

```python
type(grupo_barrio)
```

```python
grupo_barrio.groups
```

```python
for barrio, data in grupo_barrio:
    print('{} -> {}'.format(barrio, data.Valor.mean()))
```

## Estadísticas Descriptivas

```python
import pandas as pd
alumnos = pd.DataFrame({
'Nombre': ['Ary', 'Katia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
'Edad': [15, 27, 56, 32, 42, 21, 19, 35], 
'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
'Aprobado': [True, False, False, True, True, True, False, False]}, 
columns = ['Nombre', 'Edad', 'Sexo', 'Notas', 'Aprobado'])
# ¿Cómo debemos proceder para obtener un DataFrame con las notas
# promedio de los alumnos, con dos casas decimales, agrupado por su sexo?
```

```python
sexo1 = alumnos.groupby('Sexo')
sexo1 = pd.DataFrame(sexo1['Notas'].mean().round(2))
sexo1.columns = ['Notas Médias']

sexo2 = alumnos.groupby('Sexo')
sexo2 = pd.DataFrame(sexo2['Notas'].mean())
sexo2.columns = ['Notas Médias']
sexo1, sexo2
```

```python
precios = pd.DataFrame([
['Feria', 'Cebolla', 2.5], 
         ['Mercado', 'Cebolla', 1.99], 
         ['Supermercado', 'Cebolla', 1.69], 
         ['Feria', 'Tomate', 4], 
         ['Mercado', 'Tomate', 3.29], 
         ['Supermercado', 'Tomate', 2.99], 
         ['Feria', 'Papa', 4.2], 
         ['Mercado', 'Papa', 3.99], 
         ['Supermercado', 'Papa', 3.69]], 
columns = ['Local', 'Producto', 'Precio'])
precios
```

```python
productos = precios.groupby('Producto')
productos.describe().round(2)
```

```python
# Usando el DataFrame anterior (precios), qué código se necesitaría para generar la siguiente visualización:

# Producto	Promedio	Desviación Estándar	Mínimo	Máximo
# Papa	     3.96	            0.26	     3.69     4.2
# Cebolla	 2.06	            0.41	     1.69	  2.5
# Tomate	 3.43	            0.52    	 2.99	  4.0
```

```python
estadisticas1 = ['mean', 'std', 'min', 'max']
nombres = {'mean': 'Promedio', 'std': 'Desviación Estándar', 
    'min': 'Mínimo', 'max': 'Máximo'}
P0=productos['Precio'].aggregate(estadisticas1).rename(columns = nombres).round(2)
P1=productos['Precio'].agg(['mean', 'std']).round(2)

P0,P1
```

```python

estadisticas2 = ['mean', 'std', 'min', 'max']
nombres = {'mean': 'Promedio', 'std': 'Desviación Estándar', 
    'min': 'Mínimo', 'max': 'Máximo'}
productos['Precio'].aggregate(estadisticas2).rename(columns = nombres)

```
