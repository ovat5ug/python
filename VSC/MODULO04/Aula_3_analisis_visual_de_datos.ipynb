```python
import pandas as pd
#pip install Seaborn biblioteca para analisis estadisticos por medio de grafiscos de dise침os
#variados y coloridos
import seaborn as sns
tarjetas=pd.read_csv('credit_card_ESP_rangoEdad.csv')
tarjetas['iu'] = tarjetas['valor_factura'] / tarjetas['limite']
#"iu" = indice de uso: es el porcentae que se le da a una persona por el uso de la tarjeta
#en por centaje por cien
```

```python
tarjetas.head()
```

```python
sns.countplot(x='cuenta_corriente', data= tarjetas)
#nuestra variables categoricas "cuenta_corriente"
```

```python
sns.countplot(x='cuenta_corriente', data= tarjetas, hue='moroso', palette='coolwarm')
#nuestra variables categoricas "cuenta_corriente" y "moroso"
```

```python
sns.catplot(x='estado_civil', y='limite', data=tarjetas, hue='moroso', dodge=True)
```

```python
sns.swarmplot(x='escolaridad', y='iu', data=tarjetas, palette="deep")
```

```python
sns.boxplot(x='escolaridad', y='iu', data=tarjetas, hue='moroso', palette='bone')
#linea inferior significa limite minimo
#linea superior significa limite maximo
#la linea de en medio significa la mediaana
#si la mediana esta por debajo es que es menor al 50% y el limete inferior de esta es el 25%
#si la mediana esta por encima es que es mayor al 50% y el limete superior de esta es el 75%
```

```python
sns.violinplot(x='escolaridad', y='iu', data=tarjetas, hue='moroso')
```

```python
bins = [20, 30, 40, 50, 100]
nombres = ['20-30', '30-40', '40-50', '50+']

tarjetas['rango_edad'] = pd.cut(tarjetas['edad'], bins, labels=nombres)
```
