# JointPlot

```python
import pandas as pd
#pip install Seaborn biblioteca para analisis estadisticos por medio de grafiscos de diseños
#variados y coloridos
import seaborn as sns
tarjetas=pd.read_csv('credit_card_ESP_completo.csv')
```

```python
tarjetas.head()
```

```python
sns.jointplot(x='edad', y='limite', data=tarjetas)
```

```python
sns.jointplot(x='edad', y='limite', data=tarjetas, kind='kde', hue='moroso')
```

```python
sns.jointplot(x='edad', y='iu', data=tarjetas, kind='scatter')
#Nos han pedido interpretar las variables edad e índice de uso 
#(‘iu’) en nuestra base de datos
```

# Pairplot

```python
tarjetas.describe()
```

```python
sns.pairplot(data=tarjetas)
#para hacer un vistazo general de nuestra base de datos por vez primera
#y saber como los valores de las variables numericas estan relacionadas entre si
```

```python
sns.pairplot(data=tarjetas, hue='escolaridad', palette='cividis')
#para hacer un vistazo general de nuestra base de datos por vez primera
#y saber como los valores de las variables numericas estan relacionadas entre si
#en este caso analizando sobre escolaridad todas las variables numericas, porque el
#parámetro "hue" nos sirve para analizar nuestros gráficos a partir de otra variable
#con el parámetro kind, las opciones que éste parámetro nos permite son:
#(“scatter”, “kde”, “hist”, “hex”, “reg”, “resid” )
```
