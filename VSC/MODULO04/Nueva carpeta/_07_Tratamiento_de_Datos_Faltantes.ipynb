## Tratamiento de Datos Faltantes (continuación)

```python
datos[datos['Mantenimiento'].isnull()].shape[0]#saber cuales registros tiene valores nulos
```

```python
#filtrando por departamento en donde mantenimiento sea nulo
seleccion = (datos['Tipo'] == 'Departamento') & (datos['Mantenimiento'].isnull())
```

```python
A = datos.shape[0]
datos = datos[~seleccion]#la vrigulilla"~" al anteponerlo a "seleccion" 
B = datos.shape[0]       #trae todo lo contrario a "selecion" es decir
A - B                    #de esta forma no traera todo el contenido que no sea nulo
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos = datos.fillna({'Mantenimiento': 0, 'Impuesto': 0})#fillna rellena los nulos por cero
                    #valor de busque:valor que tendra cuando encuentre nulo
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]#buscabdo si aun hay nulos en mantenimiento
```

```python
datos[datos['Impuesto'].isnull()].shape[0]#buscabdo si aun hay nulos en mantenimiento
```

```python
inmuebles= pd.DataFrame([['Departamento', None, 970, 68], 
                        ['Departamento', 2000, 878, 112], 
                        ['Casa', 5000, None, 500], 
                        ['Departamento', None, 1010, 170], 
                        ['Departamento', 1500, 850, None], 
                        ['Casa', None, None, None], 
                        ['Departamento', 2000, 878, None], 
                        ['Departamento', 1550, None, 228], 
                        ['Departamento', 2500, 880, 195]], 
                        columns = ['Tipo', 'Valor', 'Mantenimiento', 'Impuesto'])
```

```python
# Elimina los registros que tengan la columna Valor nula:
inmuebles.dropna(subset = ['Valor'], inplace = True)
inmuebles
```

```python
#Elimina los inmuebles de tipo Departamento que tengan la columna Mantenimiento nula:
seleccion = (inmuebles['Tipo'] == 'Departamento') & (inmuebles['Mantenimiento'].isnull())
inmuebles = inmuebles[~seleccion]
inmuebles
```

```python
#Reemplaza los valores nulos que quedan en las variables Mantenimiento y Impuesto por cero:
inmuebles = inmuebles.fillna({'Mantenimiento': 0, 'Impuesto': 0})
inmuebles
```

```python
#Reconstruye el índice del DataFrame resultante:
inmuebles.index = range(inmuebles.shape[0])
inmuebles
```
