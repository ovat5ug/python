## Creando Nuevas Variables

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial_sinIndices.csv', sep = ';')
```

```python
datos.head(10)
```

```python
datos['Valor Bruto'] = datos['Valor'] + datos['Mantenimiento'] + datos['Impuesto']
#añade el campo "valor bruto" y calcula el valor de los campos seleccionados
```

```python
datos.head(10)
```

```python
datos['Valor m2'] = datos['Valor'] / datos['Area']
```

```python
datos.head(10)
```

```python
datos['Valor m2'] = datos['Valor m2'].round(2)
#".round(2)" redondea a dos decimales el valor
```

```python
datos.head(10)
```

```python
datos['Valor Bruto m2'] = (datos['Valor Bruto'] / datos['Area']).round(2)
#".round(2)" redondea a dos decimales el valor
```

```python
datos.head(10)
```

```python
Home = ['Casa', 'Casa en condominio', 'Casa de villa']
```

```python
datos['Tipo Agrupado'] = datos['Tipo'].apply(lambda x: 'Casa' if x in Home else 'Departamento')
#coloca "Casa si en "X" existen los valores de la variable "Home" sino pon "Departamento"
```

```python
datos
```
